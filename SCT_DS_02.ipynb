```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = 'gender_submission.csv'  # Replace with the correct file path
data = pd.read_csv(file_path)
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data.describe()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='Survived', data=data, palette='pastel')
plt.title('Survival Distribution')
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.show()
```

```python
survival_rate = data['Survived'].value_counts(normalize=True) * 100
print("\nSurvival Rate (in %):")
print(survival_rate)
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='PassengerId', y='Survived', data=data, color='skyblue', alpha=0.7)
plt.title('Passenger Survival by ID')
plt.xlabel('Passenger ID')
plt.ylabel('Survived (0 = No, 1 = Yes)')
plt.grid(alpha=0.3)
plt.show()
```
